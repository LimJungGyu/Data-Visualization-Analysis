* 리스트 생성

````python
a1= [2,5,6,9]
a2= [[3.5, 4.0, 6.5],[0.4, 0.9, 4.7]]
a3= [[[7,9,2],
      [4,0,8],
      [1,3,6],
      [5,8,4] ],
    [ [1,4,3],
      [0,0,0],
      [0,0,9],
      [0,0,7] ]]

#(4,3,2)     
a4=[ [[7,1],[9,4],[2,3]], [[4,0],[0,0],[8,0]], [[1,0],[3,0],[6,9]], [[5,0],[8,0],[4,7]] ]
#(3,2,4)

a5=[[[7,4,1,5],[1,0,0,0]], [[9,0,3,8],[4,0,0,0]], [[2,8,6,4],[3,0,9,7]]]
````

* `.shape`: 배열의 shape 확인

### **(2) 배열 초기화**




#### **1. `zeros()`**

배열의 모든 요소를 0으로 초기화 합니다.


```
배열 = np.zeros(shape)
```



````python
a= np.zeros(3)
a, a.shape
````

#### **2. `ones()`**

배열의 모든 요소를 1 로 초기화 합니다.
```
배열 = np.ones(shape)
```

````python
 a= np.ones(3)
 a, a.shape
````

#### **3. `full()`**

배열의 모든 요소를 지정 값으로 초기화 합니다.
```
배열 = np.full(shape, 지정값)
```

````python
 a= np.full(3,5)
 a, a.shape
````

### **(3) 배열 생성**




#### **1. `eye()`**

단위행렬 (identity matrix)를 생성합니다. 단위 행렬이란 주대각선의 원소가 모두 1이고 나머지 원소는 모두 0인 정사각 행렬입니다.



```
배열 = np.eye(shape)
```



````python
a = np.eye(3)
a, a.shape
````

#### **2. `empty()`**

초기화되지 않은 배열을 생성합니다. 초기화 되지 않은 배열은 생성이 빠릅니다.

```
배열 = np.empty(정수)
```

````python
a = np.empty(3)
a, a.shape
````

#### **3. `_like()`**

특정 배열과 같은 shape가 같은 행렬을 생성합니다.

  * `np.zeros_like()`
  * `np.ones_like()`
  * `np.full_like()`
  * `np.empty_like()`

```
배열2  = np.()_like(배열1)
```


````python
b=np.zeros_like(a)
b, b.shape
````

### **(3) 생성한 값으로 배열 생성**




#### **1. `arange()`**

정수 범위의 배열을 생성합니다.

```
배열 = np.arange(start, end, step)
```


````python
a = np.arange(1,10,2)  #디폴트값은 입력안해도된다(0,10,1)
a,  a.shape       #1차원 배열밖에 안된다 
````

#### **2. `linspace()`**

범위 내 균등 간격을 가지는 배열을 생성합니다.

```
배열  = np.linspace(start, end, bins)
```


````python
a = np.linspace(1,11,20)
a, a.shape
````

#### **3. `logspace()`**

범위 내 균등 간격을 가지는 로그 스케일의 배열을 생성합니다.

```
배열2  = np.logspace(start, end, bins)
```


````python
a= np.logspace(2,125,10)
a, a.shape
````

| `randn` | 표준정규분포를 따르는 표본 추출 |
| `binomial` | 이항분포를 따르는 표본 추출 |

#### **1. `rnadom.random()`**

랜덤 값을 지니는 배열을 생성합니다.



```
배열 = np.random.random(shape)
```



````python
a=np.random.random((2,3))
a,a.shape           #0부터 1사이의float값
````

#### **2. `random.randint()`**

일정 범위의 랜덤 정수값을 가지는 배열을 생성합니다.

```
배열 = np.random.randint(범위1, 범위2, shape)
```

````python
a= np.random.randint(0,10,shape)
a, a.shape
````

#### **3. `random.normal()`**
![](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/350px-Standard_deviation_diagram.svg.png)





정규분포 (normal distribution) 을 따르는 랜덤 값 배열을 생성합니다. 정규분포의 평균은 0, 표준편차는 1입니다.

```
배열 = np.random.normal(평균, 표준편차, shape)
```

````python
 a=np.random.normal(2,1,(2,3))              #(0,1,shape)를 주면 표준정규분포가 된다
 a, a.shape
````

#### **4. `random.rand()`**

균등분포 (unifrom distribution)을 따르는 랜덤한 수의 배열 생성합니다.

```
배열 = np.random.rand(shape)
```

````python
a= np.random.rand(10,2) #괄호없이 shape
a, a.shape
````

#### **5. `random.randn()`**

표준 정규 분포(standard normal distribution)를 따르는 랜덤한 수의 배열을 생성합니다.

```
배열 = np.random.randn(shape)
```

````python
a = np.random.randn(3,2)
a , a.shape
````

### **(5) 데이터 타입**

| 함수 | 설명 |
|--|--|
| `bool_` | 불리언(Boolean) 값. True, False 값을 지님|
| `int_` | 정수 (integer)|
| `intc` |  c 언어에서 사용되는 `int` 와 동일 |
| `intp` | 인덱싱에 사용되는 정수 |
| `int8` | 바이트(Byte)(-128 ~ 127)|
| `int16` | 정수 |
| `int32` | 정수|
| `int64` | 정수 |
| `uint8` | 부호없는 정수 (0~ 255) |
| `uint16` | 부호없는 정수  |
| `uint32` | 부호없는 정수  |
| `uint64` | 부호없는 정수 |
| `float16` | 반정밀 부동 소수점(Yalf precision float) |
| `float32` | 단정밀 부동 소수점(Single precision float)|
| `float64` | 배정밀 부동 소수점(Double precision float)|
| `float_` | =`float64` |

배열 = np.생성(dtype=타입)

````python
a = np.random.random((2,3))
print(a,a.shape)
a = np.array(a, dtype='float16')
print(a, a.shape)   #보통은 디폴트 값인 float32를 쓴다
````
