#### **1. `T` 전치 (transopose)**


<p align='center'>
<img src=https://www.mybluelinux.com/img/post/posts/0123/NumPy-Illustrated-30.webp width=750>
</p>

```
배열2 = 배열1.T
```








````python
a = np.arange(1,7).reshape(2,3)
print(a,a.shape)
a_t =a.T
print(a_t,a_t.shape)

a_t2=np.transpose(a)  ##두가지방법 x.T // np.transpose(x)
print(a_t2,a_t2.shape)
````

````python
b = np.arange(1,4).reshape(1,3)
print(b)

b_t=b.T
print(b_t,b_t.shape)

b_t2=np.transpose(b)
print(b_t2,b_t2.shape)
````

````python
c = np.arange(1,4).reshape(3,)
print(c,c.shape)

c_t=c.T
print(c_t,c_t.shape)

c_t2=np.transpose(c)
print(c_t2,c_t2.shape)
````

#### **2. `swap()`**

<p align='center'>
<img src=https://www.mybluelinux.com/img/post/posts/0123/NumPy-Illustrated-58.webp width=650>
</p>

```
배열2 = 배열1.swapaxes(축1, 축2)
```







````python
 a = np.arange(1,25).reshape(2,4,3)
 print((a,a.shape))

 a=np.swapaxes(a,0,1)          #a의 shape가 바뀐다 
 print(a,a.shape)
````

````python
 a = np.arange(1,25).reshape(2,4,3)
 print((a,a.shape))

 a=np.swapaxes(a,0,2)          #a의 shape가 바뀐다 
 print(a,a.shape)
````

#### **3. `reshape()`**


```
배열2 = 배열1.reshape(shape)
```

```
a =  np.zeros((2,3,5))
print(a, a.shape)
a_reshape = a.reshape((3,2,5))
print(a_reshape, a_reshape.shape)
```




````python
a =  np.arange(30).reshape(2,3,5)
print(a, a.shape)
a_reshape = a.reshape((3,2,5))
a_swap=np.swapaxes(a,0,1)
print(a_reshape, a_reshape.shape)
print(a_swap,a_swap.shape)
````

#### **4. `resize()`**

`resize()` 함수는 shape 크기를 증가시킬수도 있고 감소 시킬수도 있습니다. 다만 증가시킬 때는 빈 공간을 0으로 채우고 감소할 때는 사이즈에 포함되지 않는 값은 삭제됩니다.


```
배열1.resize(shape)
```

```
a =  np.arange(1,16)
print(a, a.shape)
a.resize((3,5))
print(a, a.shape)
a.resize((3,8))
print(a, a.shape)
a.resize((3,2))
print(a, a.shape)
```



````python
a =  np.arange(1,16) ###reshape는 개수와 차원이 맞아야되지만 resize는 알아서 채우고(0) 줄여준다(끝값 삭제)
print(a, a.shape)
a.resize((3,5))
print(a, a.shape)
a.resize((3,8))
print(a, a.shape)
a.resize((3,2))
print(a, a.shape)
````

#### **1. `append()`**



<p align='center'>
<img src=https://www.mybluelinux.com/img/post/posts/0123/NumPy-Illustrated-39.webp width=650>

`axis` 지정이 없다면 1차원 배열 형태로 변형되어 결합합니다. `axis`를 n으로 지정시 `shape[n]`을 제외한 나머지 shape는 같아야 합니다.

```
배열3 = np.append(배열1, 배열2)
```






````python
a=np.arange(1,16).reshape(3,5)
print(a,a.shape)
b= np.ones(5).reshape(1,5)
print(b,b.shape)
c=np.append(a,b,axis=0)
print(c,c.shape)
````

````python
a=np.arange(1,16).reshape(3,5)
print(a,a.shape)
b= np.zeros((3,2))
print(b,b.shape)
c=np.append(a,b,axis=1)
print(c,c.shape)
````

### **(3) 배열 쌓기**
---

<p align='center'>
<img src=https://www.mybluelinux.com/img/post/posts/0123/NumPy-Illustrated-55.webp width=600>
</p>




````python
a = np.arange(1,25).reshape(3,4,2)
print("a\n",a,a.shape)
b = np.arange(1,17).reshape(2,4,2)
print("b\n",b,b.shape)
c = np.arange(1,13).reshape(3,2,2)
print("c\n",c,c.shape)
d = np.arange(1,25).reshape(3,4,2)
print("d\n",d,d.shape)
````

####**1. `hstack()`**

수평으로 배열을 쌓습니다.(axis=1)

```
배열3 = np.hstack([배열1, 배열2])
```


````python
new_arr = np.hstack([a,c])
print(a.shape,c.shape)
print(new_arr,new_arr.shape)
````

####**2. `vstack()`**

수직으로 배열을 쌓습니다.(axis=0)
```
배열3 = np.vstack([배열1, 배열2])
```


````python
new_arr = np.vstack([a,b])
print(a.shape,c.shape)
print(new_arr,new_arr.shape)
````

####**3. `dstack()`**

새로운 축으로 배열을 쌓습니다.(axis=2)
```
배열3 = np.dstack([배열1, 배열2])
```


````python
a0 =np.zeros((3,2))
a1=np.ones((3,2))

new_arr = np.dstack([a0,a1])
print(a0.shape,a1.shape)
print(new_arr,new_arr.shape)
````

####**4. `stack()`**

새로운 축으로 배열을 쌓습니다.
```
배열3 = np.stack([배열1, 배열2])
```

````python
a0 =np.zeros((3,2))
a1=np.ones((3,2))

new_arr = np.stack([a0,a1])
print(a0.shape,a1.shape)
print(new_arr,new_arr.shape)
````

### **(4) 배열 분할**
---

<p align='center'>
<img src=https://www.mybluelinux.com/img/post/posts/0123/NumPy-Illustrated-35.webp  width=650>
</p>



````python
a = np.arange(1,13).reshape(3,4)
b = np.arange(1,9).reshape(2,4)
c = np.arange(1,7).reshape(3,2)
print(a.shape,b.shape,c.shape)
````

#### **1. `split()`**


```
배열n1, .. 배열nm = np.split(배열, [n1, .. nm], axis=축)
```










#### **2. `vsplit()`**

수직 방향으로 배열을 분할합니다.

```
배열n1, .. 배열nm = np.vsplit(배열, [n1, .. nm]
```




````python
print(a.shape,b.shape)
new_arr=np.vstack((a,b))
print(new_arr,new_arr.shape)
new_arr0, new_arr1 =np.vsplit(new_arr,[3]) 
print(new_arr0)
print(new_arr1)
````

#### **3. `hsplit()`**

수평방향으로 축을 분할합니다.

```
배열n1, .. 배열nm = np.hsplit(배열, [n1, .. nm]
```

````python
print(a.shape,c.shape)
new_arr=np.hstack((a,c))
print(new_arr,new_arr.shape)
new_arr0, new_arr1 =np.hsplit(new_arr,[3]) 
print(new_arr0,new_arr0.shape)
print(new_arr1,new_arr1.shape)
````
