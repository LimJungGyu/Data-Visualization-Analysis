### **(4) 비교연산**

비교 연산자를 이용하면 boolean 값으로 이루어진 배열을 반환합니다.





#### **1. 값 비교**


|연산자|비교 범용 함수|
|--|--|
|`==`|`np.equal`|
|`!=|`np.not_equal`|
|`<`| `np.less`|
|'<=`|`np.less_equal`|
|`>`| `np.greater`|
|'>=`|`np.greater_equal`|

````python
print(a,b)
mask = a==b
mask = a>2
print(mask)
print(a[mask],b[mask])

x= np.arange(5)
y= np.array(["a","a","b","b","a"])
print(x)
print(y)
mask_a = y =="a"
mask_b = y =="b"
print("---------mask a---------")
print(x[mask_a])
print(y[mask_a])
print("---------mask b---------")
print(x[mask_b])
print(y[mask_b])
````


#### **2. 예외값 비교**



|비교 범용 함수|설명|
|--|--|
|`np.isinf`|`inf` 값이면 `True`, 아니면 `False`|
|`np.isfinite`|`inf, nan` 값이면 `False`, 아니면 `True`|
|`np.isnan`|`nan` 값이면 `True`, 아니면 `False`|

````python
a = np.array([1,1,1,1,1,1,np.nan,np.inf])
print(np.isfinite(a))
###마스크 씌워서 숫자값만 출력하기
mask = np.isfinite(a)  
print(a[mask])

a = np.array([1,1,1,1,1,1,np.nan,np.inf])
print(np.isinf(a))
###마스크 씌워서 숫자값만 출력하기
mask = np.isinf(a)  
print(a[mask])

a = np.array([1,1,1,1,1,1,np.nan,np.inf])
print(np.isnan(a))
###마스크 씌워서 숫자값만 출력하기
mask = np.isnan(a)  
print(a[mask])
````
